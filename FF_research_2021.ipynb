```python
#import functions
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Pull in Fantasy Pros ADP File and read first 5 lines
csv ="fp_adp.csv"
df = pd.read_csv(csv)
df.head()
```

```python
#Change player name header to 'Player' for merge
df2 = df.rename(columns={"PLAYER NAME": "Player"})
df2.head()
```

```python
#Pull in NFL rushing Data from The Football Database - Still need to edit

URL = 'https://www.footballdb.com/statistics/nfl/player-stats/rushing/2020'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
rb_table = soup.find("table", attrs={"class": "statistics"})
rb_table_data = rb_table.tbody.find_all("td")  # contains 2 rows
```

```python
#2020 Rushing Stats
rb = "2021_rushing.csv"
df_rb = pd.read_csv(rb)
df_rb.dtypes
```

```python
#join dataframes together
df_merge = pd.merge(df2,df_rb, on='Player')
df_merge.head()
```

```python
#Create Fantasy Points Scored Column
df_merge['Rush_FF_Points'] = df_merge['Yds']/10 + df_merge['TD']*6
df_merge.head(10)
df_merge.sort_values(by='FF_Points', ascending=False).head(20)
```

```python
csv2 = '2020_rec_1.csv'
wr_1 = pd.read_csv(csv2)
wr_1.head()
merge_2 = pd.merge(df_merge, wr_1, on='Player')
merge_2.head()
merge_2.columns
merge_3 = merge_2
```

```python
merge_4 =merge_3.rename(columns={'Yds_x':'Yds_rush', 'Avg_x': 'Avg_rush', 'YPG_x': 'YPG_rush', 'Lg_x': 'Lg_rush', 'TD_x': 'Rush TD', 'FD_x': 'FD_rush', 
        'Yds_y': 'Yds_Rec' , 'Avg_y': 'Avg_Rec', 'YPG_y' : 'YPG_Rec',
       'Lg_y': 'Lg_Rec', 'TD_y': 'Rec TD', 'FD_y': 'FD_Rec', 'Tar': 'Targets'})

merge_4.columns
```

```python
merge_5 = merge_4.drop(columns=['Team_y','FF_Points','Gms_y', 'BYE WEEK'])
merge_5.head()
```

```python
merge_5['FF_Rec'] = (merge_5['Rec'] * .5) + (merge_5['Yds_Rec']/10) + merge_5['Rec TD'] * 6
merge_5.head()
```

```python
#Top Receiving Fantasy Points Scored by RBs
df_5 = merge_5
df_5.sort_values(by='FF_Rec', ascending=False)
df_5['Rec_Rank'] = df_5['FF_Rec'].rank()
df_5.sort_values(by='Rec_Rank',ascending=False)
```

```python
df_5['Total_FP'] = df_5['Rush_FF_Points'] + df_5['FF_Rec']
df_5.head()
```
